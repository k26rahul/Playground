# Deriving the joint pmf using the marginal and conditional pmf



```python
X = [1, 2, 3, 4, 5, 6]
Y = [0, 1, 2, 3, 4, 5, 6]
```

```python
pmf_matrix = np.zeros((len(X), len(Y)))
```

```python
def pmf_XY(t, _t):
  return 1/6 * comb(t, _t) * 1/2 ** t
```

```python
for i, x_val in enumerate(X):
  for j, y_val in enumerate(Y):
    pmf_matrix[i, j] = pmf_XY(x_val, y_val)
```

# `Colormaps` available in Matplotlib

"Viridis" is one of the built-in colormaps available in Matplotlib. It is a perceptually uniform colormap designed to be visually pleasing and friendly to colorblind viewers. It ranges from a deep blue color to a bright yellow, making it suitable for representing sequential data.

Some common colormaps include:

1. **'jet'**: This is a rainbow colormap that ranges from blue to red. It was widely used in the past but is now discouraged due to perceptual issues.
  
2. **'coolwarm'**: This colormap ranges from cool colors (blue) to warm colors (red).

3. **'cividis'**: Similar to 'viridis', this colormap is designed to be perceptually uniform and is suitable for colorblind viewers.

4. **'inferno'**: This colormap ranges from black to yellow, with intermediate oranges and reds.

5. **'plasma'**: This colormap ranges from deep purple to bright yellow.

6. **'magma'**: This colormap ranges from black to bright yellow through shades of purple.

7. **'summer'**: This colormap ranges from green to yellow.

8. **'cool'**: This colormap ranges from cyan to magenta.

9. **'hot'**: This colormap ranges from black to red through shades of yellow.

10. **'gray'**: This colormap is a linear grayscale colormap ranging from black to white.

# The `origin` and `interpolation` parameters

The `origin` and `interpolation` parameters in the `imshow()` function of Matplotlib allow you to control the interpretation and display of the image data.

1. **`origin`**: This parameter specifies where the origin of the image is located. It affects the interpretation of the x and y coordinates.

   Possible values:
   - `'upper'`: The origin is at the upper left corner of the image (default for Matplotlib). In this case, the row index increases downward, and the column index increases to the right.
   - `'lower'`: The origin is at the lower left corner of the image. In this case, the row index increases upward, and the column index increases to the right.

2. **`interpolation`**: This parameter controls how the image data is [interpolated](https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html) when it is displayed.

   Possible values:
   - `'nearest'`: Uses the nearest neighbor interpolation. Each pixel in the image is displayed using the value of the nearest pixel in the original data.
   - `'bilinear'`: Uses bilinear interpolation. This is a linear interpolation method that considers the values of the four nearest pixels to calculate the value of each pixel in the displayed image.
   - `'bicubic'`: Uses bicubic interpolation. This is a cubic interpolation method that considers the values of the sixteen nearest pixels to calculate the value of each pixel in the displayed image.
   - `'spline16'`, `'spline36'`, `'hanning'`, `'hamming'`, `'hermite'`, `'kaiser'`, `'quadric'`, `'catrom'`, `'gaussian'`, `'bessel'`, `'mitchell'`, `'sinc'`, `'lanczos'`: Other interpolation methods are also available, each with its own characteristics.
