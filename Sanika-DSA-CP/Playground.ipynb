# I have a long tail

```python
num = 12.3214

str(num)
str(num).split('.')
```

```python
n_left = 0

for digit in str(num):
  if digit == '.':
    break
  n_left += 1

n_right = len(str(num)) - n_left

if n_right > n_left:
  print('I have a long tail')
```

```python
L = str(num).split('.')

n_left = len(L[0])
n_right = len(L[1])

if n_right > n_left:
  print('I have a long tail')
```
